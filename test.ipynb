```python
from urllib.parse import urlparse
```

```python
parsed_url = urlparse("https://www.linkedin.com/in/zackspear/")
paths = [path for path in parsed_url.path.strip().split("/") if path]
print(
    parsed_url.scheme,
    parsed_url.netloc,
    parsed_url.path,
    paths
)
parsed_url.scheme != "https" or parsed_url.netloc != "www.linkedin.com" or "/in/" not in parsed_url.path or len(paths) != 2
```
